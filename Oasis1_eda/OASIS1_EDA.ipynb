sales,profit and discount by sub_category

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
menu_df = pd.read_csv("Downloads/supermart.csv", encoding='unicode_escape')

# Group by 'Sub Category' and calculate the sum of 'Sales', 'Profit', and 'Discount', then sort by 'Sales'
sales_by_subcategory = menu_df.groupby('Sub Category')[['Sales', 'Profit', 'Discount']].sum().sort_values(by='Sales', ascending=False).reset_index()

# Plotting
sales_by_subcategory.plot(x='Sub Category', y=['Sales', 'Profit', 'Discount'], kind='bar', figsize=(12, 8))
plt.title('Sales, Profit, and Discount by Sub Category')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import squarify
import pandas as pd

# Assuming menu_df is already loaded
category_sales = menu_df.groupby(['Category', 'Sub Category']).agg({'Sales': 'sum'}).reset_index()

# Prepare labels and sizes for the treemap
labels = category_sales['Category'] + ' - ' + category_sales['Sub Category']
sizes = category_sales['Sales']

# Plotting the treemap
plt.figure(figsize=(10, 8))
squarify.plot(
    sizes=sizes, 
    label=labels, 
    alpha=0.7, 
    color=plt.cm.viridis(sizes / max(sizes)), 
    text_kwargs={'fontsize': 10}
)

plt.title("Treemap of Sales by Category and Sub Category")
plt.axis('off')
plt.show()
```

interactive visualisations with bokeh

1.interactive line chart for sales and profit over time

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
import pandas as pd

# Example time series data for testing
menu_df = pd.DataFrame({
    'Order Date': pd.date_range(start='2024-01-01', periods=10, freq='D'),
    'Sales': [100, 120, 130, 115, 125, 135, 145, 155, 165, 175],
    'Profit': [20, 25, 30, 28, 27, 33, 37, 40, 43, 45]
})

# Group data by 'Order Date' and calculate the sum of 'Sales' and 'Profit'
time_series_data = menu_df.groupby('Order Date')[['Sales', 'Profit']].sum().reset_index()

output_notebook()

# Create ColumnDataSource from time_series_data
source = ColumnDataSource(time_series_data)

# Create a figure for the line plot
p = figure(
    title='Sales and Profit Over Time',
    x_axis_label='Order Date',
    y_axis_label='Amount',
    x_axis_type='datetime',
    height=400,
    width=800
)

# Plot Sales and Profit lines
p.line('Order Date', 'Sales', source=source, legend_label='Sales', line_width=2, color='blue')
p.line('Order Date', 'Profit', source=source, legend_label='Profit', line_width=2, color='green')

# Customize the legend
p.legend.location = "top_left"
p.legend.click_policy = "hide"

# Show the plot
show(p)
```

```python
import pandas as pd
from bokeh.plotting import figure,show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
menu_df = pd.read_csv("Downloads/supermart.csv",encoding='unicode_escape')



time_series_data=menu_df.groupby('Order Date')[['Sales','Profit']].sum().reset_index()
output_notebook()

source=ColumnDataSource(time_series_data)
p = figure(
    title='Sales and Profit Over Time',
    x_axis_label='Order Date',
    y_axis_label='Amount',
    x_axis_type='datetime',
    height=400,
    width=800
)
p.line('Order Date', 'Sales', source=source, legend_label='Sales', line_width=2, color='blue')
p.line('Order Date', 'Profit', source=source, legend_label='Profit', line_width=2, color='green')

# Customize the legend
p.legend.location = "top_left"
p.legend.click_policy = "hide"

# Show the plot
show(p)
```
